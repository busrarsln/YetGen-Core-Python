```python
#Ups, programımıza hoş geldiniz! ^^
isim = input("İsminiz: ")
yas = int(input("Yaşınız: "))
eposta = input("E-posta adresiniz: ")
print(f" {isim} Cep Muhasebecisi'ne hoş geldiniz! Cep Muhasebecisi sayesinde gelir ve giderlerinizi hesaplayabilir, harcama hedefinizi bu verilerle inceleyebilir ve enflasyon etkisini görebilirsiniz.")




while True:
  
   print("1. Aylık enflasyon oranına göre lüks giderlerimi hesapla.")
   print("2. Gelir ve gider hesabını yap, raporla.")
   print("3. Hedeflenen harcama bilgilerimle giderlerimi karşılaştır, raporla.")
   print("4. Çıkış")


   secim = input("Seçiminiz: ")


   if secim == "1":
       print("Lütfen hesaplamanın yapılabilmesi için gerekli bilgileri giriniz!")
       print("Aylık enflasyon oranına göre lüks giderler hesaplanıyor...")
       try:
           sabit_gelir =float (input("Sabit geliriniz: "))
           degisken_gelir = float(input("Değişken geliriniz: "))
           hedeflenen_harcama= float(input("Bu ay harcamak istediğiniz miktarı giriniz: "))
       except ValueError:
           print("Hatalı giriş! Lütfen sadece sayısal değerler girin.")
       toplam_gelir = sabit_gelir + degisken_gelir


       class SabitGiderler:
          def __init__(self):
               try:
                  self.kira = float(input("Kira gideri: "))
                  self.elektrik = float(input("Elektrik gideri: "))
                  self.su = float(input("Su gideri: "))
                  self.internet = float(input("İnternet gideri: "))
                  self.dogalgaz = float(input("Doğalgaz gideri: "))
                  self.ulasım = float(input("Ulaşım harcalamarınızın gideri: "))
                  self.digerzorunlu = float(input("Diğer zorunlu giderleriniz: "))
               except ValueError:
                   print("Hatalı giriş! Lütfen sadece sayısal değerler girin.")
       sabit_giderler = SabitGiderler()


       class LuksGiderler:
           def __init__(self):
               try:
                   self.tatil = float(input("Tatil gideri: "))
                   self.restoran = float(input("Restoran gider: "))
                   self.hobi = float(input("Hobi gideri: "))
                   self.giyim = float(input("Giyim gideri: "))
                   self.digerluks = float(input("Diğer lüks giderleriniz: "))
               except ValueError:
                   print("Hatalı giriş! Lütfen sadece sayısal değerler girin.")




       luks_giderler = LuksGiderler()
       toplam_gider = sabit_giderler.kira + sabit_giderler.elektrik + sabit_giderler.su + sabit_giderler.internet + sabit_giderler.dogalgaz + luks_giderler.tatil + luks_giderler.restoran + luks_giderler.hobi + luks_giderler.giyim


       aylar = ['Ocak', 'Şubat', 'Mart', 'Nisan', 'Mayıs', 'Haziran', 'Temmuz', 'Ağustos', 'Eylül', 'Ekim', 'Kasım' , "Aralık"]
       aylik_enflasyon = [57.68, 55.18, 50.51, 37.72, 43.25, 48.00, 51.95, 41.25, 49.35, 50.62, 46.62, 52.12]


       ay_enflasyon_ciftleri = list(zip(aylar, aylik_enflasyon))
       ay = input("Bir ay ismi girin: ")
       enflasyon_orani = None
       for ay_enflasyon in ay_enflasyon_ciftleri:
           if ay_enflasyon[0] == ay:
               enflasyon_orani = ay_enflasyon[1]
               break
       if enflasyon_orani is not None:
           luks_giderler.tatil *= 1 + enflasyon_orani / 100
           luks_giderler.restoran *= 1 + enflasyon_orani / 100
           luks_giderler.hobi *= 1 + enflasyon_orani / 100
           luks_giderler.giyim *= 1 + enflasyon_orani / 100
           toplam_luksgider = luks_giderler.tatil + luks_giderler.restoran + luks_giderler.hobi + luks_giderler.giyim + luks_giderler.digerluks
           yuvarlanmis_topluks = round(toplam_luksgider, 2)
           toplam_gider = sabit_giderler.kira + sabit_giderler.elektrik + sabit_giderler.su + sabit_giderler.internet + sabit_giderler.dogalgaz + sabit_giderler.ulasım + sabit_giderler.digerzorunlu + luks_giderler.tatil + luks_giderler.restoran + luks_giderler.hobi + luks_giderler.giyim + luks_giderler.digerluks
           yuvarlanmis_topgid = round(toplam_gider, 2)
           print (f"Bu ayki lüks giderlerinize uygulanacak enflasyon oranı {enflasyon_orani}. Ayrıca lüks giderlerinizin enflasyonlu fiyatı {yuvarlanmis_topluks} tl'dir.Lüks giderlerinizin enflasyonlu fiyatını bilmek, bu harcalamarınızda ne kadar azaltma yapabileceğinizi gösterir.")
       else:
           ("Geçersiz ay ismi girdiniz.")
          
   elif secim == "2":
       print("Gelir ve gider hesaplanıyor...")
       artan_miktar = toplam_gelir - toplam_gider
       yuvarlanmis_artmik= round(artan_miktar, 2)
       asilan_miktar = toplam_gider - toplam_gelir
       yuvarlanmis_asmik = round(asilan_miktar, 2)
       artan_hedef = hedeflenen_harcama - toplam_gider
       round(artan_hedef, 2)
       asilan_hedef = toplam_gider - hedeflenen_harcama
       round(asilan_hedef, 2)
       if  (hedeflenen_harcama) ==  (toplam_gider):
           print("Hedef bütçenize tam tamına ulaştınız.")
       elif (hedeflenen_harcama) >  (toplam_gider):
           print(f"Hedef bütçenizden {artan_hedef} TL miktarında para arttırdınız, tebrikler!.")
       else:
           print("Hedef bütçenizin üzerindesiniz, bu ay birikim yapamazsınız.")
    
   elif secim == "3":
       print("Hedeflenen harcama ile giderler karşılaştırılıyor...")
       if  (toplam_gelir) ==  (toplam_gider):
          print("Bütün gelirinizi kullandınız, bu ay birikim yapamazsınız.")
       elif (toplam_gelir) >  (toplam_gider):
          print(f"Geliriniz giderlerinizden fazla artan paranız {yuvarlanmis_artmik} tl'dir.")
       else:
         print(f"Bu ayki gelirinizin üstüne çıktınız, kısmanız gereken bütçe miktarı {yuvarlanmis_asmik} tl' dir..")


   elif secim == "4":
       print("Programdan çıkılıyor...")
       break
   else:
       print("Geçersiz seçim. Lütfen tekrar deneyin.")
```
