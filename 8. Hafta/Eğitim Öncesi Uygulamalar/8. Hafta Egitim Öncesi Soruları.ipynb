# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
class Vehicle():
    def __init__(self,name,speed,mileage):
        self.name=name
        self.speed=speed
        self.mileage=mileage
class Bus(Vehicle):
    def __init__(self,name,speed,mileage):
        Vehicle.__init__(self,name,speed,mileage)
        
````

````python
bus2=Bus("School Volvo",180,12)
print("Name: {}, Speed: {}, Mileage: {}".format(bus2.name,bus2.speed,bus2.mileage))
````
