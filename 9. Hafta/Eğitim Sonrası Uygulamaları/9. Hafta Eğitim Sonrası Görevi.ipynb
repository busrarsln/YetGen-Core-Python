Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import random
import time

class Kumanda:
    def __init__(self, tv_durum="kapalı", tv_ses=0, kanal_listesi=["FOX"], kanal="FOX"):
        self.tv_durum=tv_durum
        self.tv_ses=tv_ses
        self.kanal_listesi=kanal_listesi
        self.kanal=kanal

    # TV açma-kapama
    def tv_ac(self):
        if self.tv_durum=="açık":
            print("TV zaten açık")
        else:
            print("TV açılıyor...")
            self.tv_durum="açık"
    def tv_kapat(self):
        if self.tv_durum=="kapalı":
            print("TV zaten kapalı")
        else:
            print("TV kapatılıyor...")
            self.tv_durum="kapalı"
    # ses ayarları
    def ses_ayarlari(self):
        while True:
            cevap=input("Sesi Azalt: '<'\nSesi Artır: '>'\nÇıkış: 'q'")
            if cevap=="<":
                if self.tv_ses!=0:
                    self.tv_ses-=1
                    print("Ses:", self.tv_ses)
            elif cevap==">":
                if self.tv_ses!=31:
                    self.tv_ses+=1
                    print("Ses:", self.tv_ses)
            else:
                print("Ses güncellendi:", self.tv_ses)
                break
    def kanal_ekle(self,kanal):
        print("Kanal ekleniyor...")
        time.sleep(1)
        self.kanal_listesi.append(kanal)
        print("Kanal eklendi.")
   
    # kanal sayısını öğrenme
    def kanal_sayisi(self):
        print("toplam",len(self.kanal_listesi), "kanal var.")
    
    # kanal değiştirme
    def kanal_degis(self):
        rastgele = random.randint(0, len(self.kanal_listesi)-1)
        self.kanal = self.kanal_listesi[rastgele]
        print("Şu anki kanal:", self.kanal)
    # TV bilgileri
    def tv_bilgisi(self):
        print("""
        Televizyon Durumu: {}
        Ses Düzeyi: {}
        Kanal Listesi: {}
        Şu anki Kanal: {}
        """.format(self.tv_durum, self.tv_ses, self.kanal_listesi, self.kanal))
kumanda=Kumanda()
print("""
Televizyon Kumandası

1. TV Aç
2. TV Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri

Çıkış için 'q' tuşuna basın.
""")

while True:
    islem = input("İşlem Seçiniz:")
    if islem == "q":
        print("Program sonlandırılıyor...")
        break
    elif islem=="1":
        kumanda.tv_ac()
    elif islem=="2":
        kumanda.tv_kapat()
    elif islem=="3":
        kumanda.ses_ayarlari()
    elif islem=="4":
        kanal_isimleri = input("Kanal isimlerini ',' ile ayırarak gir")
    elif islem=="5":
        kumanda.kanal_sayisi()
    elif islem=="6":
        kumanda.kanal_degis()
    elif islem=="7":
        kumanda.tv_bilgisi()
    else:
        pass
```
