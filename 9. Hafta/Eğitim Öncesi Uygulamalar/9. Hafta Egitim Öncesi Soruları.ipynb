# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
class vehicle:
    def __init__(self, name, speed, milage):
        self.name = name
        self.speed = speed
        self.milage = milage
        print ("vehicle nesnesi türetildi.")
    def intro(self):
        print(self.name, self.speed, self.milage)
class bus(vehicle):
    def __init__(self, name, speed, milage, capacity):
        super().__init__(name, speed, milage)
        self.capacity = capacity
        print("bus nesnesi türetildi.")
    def intro(self):
        print(self.name, self.speed, self.milage, self.capacity)
    def seat(self):
        print("The seating capacity of a bus is {}")
````

````python
bus1 = bus("Mercedes", 190, 12, 50)
bus1.intro()
bus1.seat()
````

````python
class vehicle:
    pass
class bus(vehicle):
    def seating_capacity(self, capacity = 50):
        self.capacity = capacity
        print(f"The seating capacity of a bus is {self.capacity} passengers.")
````
